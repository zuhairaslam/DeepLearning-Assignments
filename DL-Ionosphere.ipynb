```python
# Converting values of b & g
replace_values = {'label':{'g':0, 'b':1}}
```

```python
data = data.replace(replace_values)
```

```python
data = data.sample(frac=1)
```

```python
y = data.label
x = data.drop(columns='label')
```

```python
x.head()
```

```python
y
```

```python
# Spliting data into train and test set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x,y, test_size= 0.4 )
```

```python
X_train
```

```python
Y_train
```

```python
# Model Building
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
from tensorflow.keras import regularizers
```

```python
network = Sequential()
network.add(layers.Dense(16, activation='relu', kernel_regularizer = regularizers.l2(0.001),  input_shape=(X_train.shape[1],)))
network.add(layers.Dense(8, activation='relu' , kernel_regularizer = regularizers.l2(0.001)))
network.add(layers.Dense(4, activation='relu' , kernel_regularizer = regularizers.l2(0.001)))
network.add(layers.Dense(1, activation='sigmoid'))
network.compile(optimizer='adam', loss='binary_crossentropy', metrics='accuracy')
```

```python
network.fit(X_train, Y_train, epochs=100, batch_size=8)
```

```python
prediction = network.predict(X_test)
```

```python
network.evaluate(X_test, Y_test)
```
